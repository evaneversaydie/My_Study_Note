# 使用老師提供程式碼理解概念
下面程式碼是第一週老師提供程式碼，自己看完用註解打下來的理解。

```python
class MyLinkedList:
    
    
    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.val = None
        self.next = None
        
        # 用來建立基本單位:節點。注意的地方是因為不一定有header的存在(lined list本身為None)，所以設計初始值皆為None。
        # 每個節點都存了下一個節點的位置以及本身的value。
        # 如果是最尾巴=>self.next=None
        # 可能要設計方便寫這是第幾個index的機制。
   
        
    def get(self, index: int) -> int:
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        """
        if self.val == None:
            return -1
        if index == 0:
            return self.val
        p = self.next
        i = 1
        while p != None:
            if i == index:
                return p.val
            p = p.next
            i += 1
        return -1
    
    #此def是要得到位置index節點的val
    #只有上一個點知道下一個點的位置=>走訪要用while迴圈進行，走訪整條都不存在則回傳-1
    #有幾個情形需要被考慮到:index-th node不存在、self.val==None、查的節點是head
    

    def addAtHead(self, val: int) -> None:
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        """
        if self.val == None:
            self.val = val
            return
        temp = self.val
        self.val = val
        tempnode = self.next
        self.next = MyLinkedList()
        self.next.val = temp
        self.next.next = tempnode
        return
    #此def是要把值變成linked list的head 
    #如果本來head就是空的直接把值加在head.val
    #如果head有人了=>用 temp把原本的head存起來，原地更改head的val與next
    #結果:()
    #原:(self.val=?)(self.next)|(...)
    #後:(sel.val=val)(self.next = Mylinked list)|(temp=原本上一排的self.val=?)(tempnode=原本上一排的self.next)|(...)
    # line50、line 51是建立self和temp的關係

    
    def addAtTail(self, val: int) -> None:
        """
        Append a node of value val to the last element of the linked list.
        """
        if self.val == None:
            self.val = val
            return
        p = self
        while p.next != None:
            p = p.next
        p.next = MyLinkedList()
        p.next.val = val
        return
    #家道尾巴
    #一樣考慮head不存在再遍訪
    #到了最後一個再加一個節點(這裡直接把節點寫在Mylinkedlist()中)
    
    
    def addAtIndex(self, index: int, val: int) -> None:
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        """
        i = 0
        p = self
        pre = p
        if index <= 0:
            self.addAtHead(val)
            return
        while i < index:
            i += 1
            pre = p
            if p != None and p.val != None:
                p = p.next
            else:
                return
        pre.next = MyLinkedList()
        pre.next.val = val
        pre.next.next = p
        return
    #加到第index個元素A之前(A變成index+1個，位置為index)
    #遍訪到(位置index)，修改他的.next=一個val的新node，新node.next=原本的。
            

    def deleteAtIndex(self, index: int) -> None:
        """
        Delete the index-th node in the linked list, if the index is valid.
        """
        i = 0
        p = self
        if index < 0:
            return
        if index == 0:
            if self.val == None:
                return
            if self.next == None:
                self = None
                return
            self.val = self.next.val
            self.next = self.next.next
        pre = p
        while i < index:
            i += 1
            pre = p
            if pre == None:
                return
            p = p.next
        if p != None:
            pre.next = p.next
        else:
            pre.next = None
        return 

    #刪除某一個node，index是元素位置
```
